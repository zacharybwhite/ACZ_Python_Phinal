```python
#reading a CSV with confirmed cases from the Johns Hopkins University Center for Systems Science and Engineering

confirmed2 = pd.read_csv("us_confirmed_420.csv")
pd.set_option("display.max_columns", None)
confirmed2
```

```python
#The previous notebook shows the number of confirmed cases separated by date and broken up by smaller geographical areas
#We're just interested in state level, so we'll add them up by state to get total number of cases per state

confirmed_by_state = confirmed2.groupby("Province/State").sum()
confirmed_by_state2 = confirmed_by_state.drop("Diamond Princess")
confirmed_by_state2
```
